## 2. Hive中的数据库

　　Hive中数据库的概念本质上仅仅是表的一个目录或者命名空间，然而，对于具有很多组和用户的大集群来说，这是非常有用的，因为这样可以避免表命名冲突。通常会使用数据库来将生产表组织成逻辑组。
	
　　如果用户没有显式指定数据库，那么将会使用默认的数据库default；

### 2.1 创建一个数据库


```python
create database xxx
```

### 2.2 如果xxx已经存在，则会报错，以下语句可以避免报错

```python
create database if not exists xxx
```

### 2.3 查看数据库

```python
show databases
```

### 2.4 使用正则表达式筛选出需要的数据库名

```python
show databases like 'h.*' # 以字母h开头的
```

### 2.5 修改存放数据库的默认位置

```python
create database xxx location '/xxxxxx'
```

### 2.6 为数据库增加一个描述信息

```python
create database xxx comment 'xxxxxxxx'
```

### 2.7 查看数据库描述信息

```python
describe database xxx
```

### 2.8 创建一个数据库，并增加一些键-值对属性信息，同时进行查询

```python
create database xxx with dbproperties ('creator' = 'Mark Moneybags', 'date' = '2012-01-02')
describe database xxx # 该语句不显示键-值对属性信息
describe database extended xxx # 可以显示键-值对属性信息
```

### 2.9 将某个数据库设置为用户当前的工作数据库

```python
use xxx
# hive中没有一个命令可以让用户查看当前所在市哪个数据库，不过再Hive中，是可以重复使用use...命令的；
```

### 2.10 可以通过设置一个属性值来再提示符里面显示当前所在的数据库

```python
set hive.cli.print.current.db = true
```

### 2.11 删除数据库

　　如果想删除数据库，那么就必须先要删除掉该数据库中的所有表；如果某个数据库被删除了，那么其对应的目录也同时会被删除；

```python
drop database if exists xxx
# if exists 是可选的，避免没有该数据库而抛出警告；
```

　　默认情况下，Hive不允许用户删除一个包含有表的数据库的，用户要么先删除数据库中的表，然后在删除数据库；要么再删除命令的最后面加上关键字cascade，这样可以使Hive自行先删除数据库中的表；

```python
drop database if exists xxx cascade
```

## 3. 修改数据库

　　可以通过alter database命令为某个数据库的dbproperties设置键-值对属性值，来描述这个数据库的属性信息；

```python
alter database xxx set dbproperties ('edited-by' = 'Joe Dba')
```

　　数据库的其他元数据信息都是不可更改的，包括数据库名和数据库所在的目录位置；没有办法可以删除或者重置数据库属性；
